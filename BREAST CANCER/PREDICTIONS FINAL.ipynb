```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
# LOGISTIC REGRESSION
```

```python
m1 = LogisticRegression()
```

```python
X=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
m1.fit(X_train,y_train)
```

```python
pred1=m1.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred1))
```

```python
print(confusion_matrix(y_test,pred1))
```

```python
# SUPPORT VECTOR MACHINE WITHOUT GRID
```

```python
from sklearn.svm import SVC
```

```python
m2 = SVC()
```

```python
m2.fit(X_train,y_train)
```

```python
predictions2 = m2.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictions2))
```

```python
print(classification_report(y_test,predictions2))
```

```python
# SUPPORT VECTOR MACHINE USING GRID
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```

```python
# FROM THE ABOVE RESULTS WE CAN SHOW THAT SUPPORT VECTOR MACHINE USING GRID SEARCH IS THE BEST WAY TO PREDICT
```
